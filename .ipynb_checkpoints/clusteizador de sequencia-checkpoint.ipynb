```python
from Bio.Cluster import kcluster
import numpy as np
from Bio import SeqIO
import pandas as pd
```

```python
sequences = pd.DataFrame(columns = ['sequence',])
for seq_record in SeqIO.parse('../output_chloe/220829_19-39-44-ZTRGEOGRHO/input.rmdup.aligned.fa', "fasta"):
    # Take the current sequence
    if not seq_record.seq in sequences:
        sequences[str(seq_record.seq)] = seq_record.id
```

```python
matrix = np.asarray([np.frombuffer(s.encode(),'u1') for s in list(sequences.keys())])
clusterid, error, nfound = kcluster(matrix, nclusters=4)
```
